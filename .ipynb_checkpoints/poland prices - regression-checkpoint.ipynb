# Preprocessing Data

```python
'''
Description:

address - Full addres
city - Warszawa (Warsaw), Kraków (Cracow), Poznań (Poznan).
floor - The number of the floor where the apartment is located
id - id
latitude - latitude
longitude - longitude
price - Price of apartment in PLN [TARGET]
rooms - Number of rooms in the apartment
sq - Number of square meters of the apartment
year - Year of the building / apartment
'''
```

```python
import pandas as pd
 
file = ('house-prices-in-poland/Houses.csv')
df = pd.read_csv(file, encoding = 'ISO-8859-1')
 
df.head()
```

```python
df.groupby(['city'])['id'].count()
```

```python
price_df = df.drop(['Unnamed: 0','address','id'], axis=1)
# [['city', 'floor','id','latitude','longitude','price','rooms','sq','year']]
price_df.head()
```

```python
price_df.shape
```

```python
warsaw_df = price_df[price_df['city'] == 'Warszawa']
warsaw_df = warsaw_df.drop(['city'], axis=1)
warsaw_df.head()
```

```python
krakow_df = price_df[price_df['city'] == 'Kraków']
krakow_df = krakow_df.drop(['city'], axis=1)
krakow_df.head()
```

```python
poznan_df = price_df[price_df['city'] == 'Poznañ']
poznan_df = poznan_df.drop(['city'], axis=1)
poznan_df.head()
```
